## `np.linspace()`

#### What if I have been provided with count of elements along with range of the array?

We can generating equally spaced points using `linspace()`


#### `np.linspace()` requires us to provide range and count of elements:

```python
np.linspace(0, 10, 11)
```



#### But there is a minor catch here?

The `end` value is included in the array.


```python
start, end, count = 0, 100, 25
np.linspace(start, end, count)
```

## Overflow in numpy array

#### Another way np array behaves like C arrays and not Python lists

- In Python lists, number values can be **arbitrarily large or small**
- There's **usually no overflow of number values**


#### However, in C, C++ and Java, there's overflow of values

- As soon as a **number crosses the max possible** value for a data-type
- The number gets **wrapped around to a smaller value**

```python
100**10 # no overflow
```

```python
arr6 = np.array([0, 10, 100])
arr6 ** 10 # 100^10 will overflow
```
