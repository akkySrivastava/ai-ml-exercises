```python
# apply
# gender --> Male, Female (0, 1)
# ML Algos want all the features to be numerical

def encode(gender):
    if gender == "Male":
        return 0
    elif gender == "Female":
        return 1
    else:
        return gender
    
```

```python
data["gender"] = data["gender"].apply(encode)
data
```

```python
data["sum"] = data[["revenue", "budget"]].apply(np.sum, axis = 1)
```

```python
data["profit"] = data[["budget", "revenue"]].apply(np.diff, axis = 1)
data.head()
```

```python
def calProfit(x): 
    return x["revenue"] - x["budget"]
```
